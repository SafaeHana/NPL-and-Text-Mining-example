```python
import nltk
with open('C:/Users/safae/Documents/brown_corpus_ca10.txt','r')as myfile:
    data=myfile.read().replace('\n','')
```

```python
data2=data.replace("/","")
```

```python
for i, line in enumerate(data2.split('\n')):
    if i>10:
        break
    print(str(i) + ':\t' + line)
```

```python
from nltk import sent_tokenize,word_tokenize
```

```python
sent_tokenize(data2)
```

```python
for sent in sent_tokenize(data2):
    print(word_tokenize(sent))
```

```python
from nltk.corpus import stopwords
stopwords_en=stopwords.words('english')
```

```python
#tokenizing and lowercasing the words:
#The map() function is used to apply the str.lower() method to each element of the list returned by word_tokenize()
single_tokenized_lowered = list(map(str.lower,word_tokenize(data2)))
print(single_tokenized_lowered)
```

```python
stopwords_en=set(stopwords.words('english'))
print([word for word in single_tokenized_lowered if word not in stopwords_en])
```

```python
from string import punctuation
print('from string.punctuation:' , type(punctuation),punctuation)
```

```python
#The union() method is then used to combine the stopwords_en set with another set created from the punctuation string imported 
stopwords_en_withpunct= stopwords_en.union(set(punctuation))
print(stopwords_en_withpunct)
```

```python
#The resulting list contains only the words from the single_tokenized_lowered list that are not stopwords or punctuation characters, and can be used for further analysis of the text data.
print([word for word in single_tokenized_lowered if word not in stopwords_en_withpunct])
```

```python
#remove suffixes from words.
from nltk.stem import PorterStemmer
porter=PorterStemmer()
for word in single_tokenized_lowered:
    print(porter.stem(word))
```

```python
from nltk.stem import WordNetLemmatizer
wnl=WordNetLemmatizer()
```

```python
#The lemmatize() method uses the WordNet lexical database to determine  origine and correct base form of each word,example: cats=>cat , ran=>run
for word in single_tokenized_lowered:
    print(wnl.lemmatize(word))
```

```python
stop_words = set(stopwords.words('english'))
tokenized=sent_tokenize(data2)
for i in tokenized:
    wordlist = nltk.word_tokenize(i)
    wordlist =[w for w in wordlist if not w in stop_words]
    tagged = nltk.pos_tag(wordlist)
    print(tagged)
```

```python
sentences = nltk.sent_tokenize(data2)
tokenized_sentences = [nltk.word_tokenize(sentence)for sentence in sentences]
tagged_sentences=[nltk.pos_tag(sentence)for sentence in tokenized_sentences]
#he program groups the words into phrases and uses nltk.ne_chunk_sents() to identify named entities. A named entity is a word or group of words that refers to a person, place, or organization.
chunked_sentences = nltk.ne_chunk_sents(tagged_sentences, binary=True)

def extract_entity_names(t):
    entity_names = []
    if hasattr(t,'label') and t.label:
        if t.label()=='NE':
            entity_names.append(' '.join([child[0]for child in t]))
        else:
            for child in t:
                entity_names.extend(extract_entity_names(child))
    return entity_names
entity_names = []
for tree in chunked_sentences:
    entity_names.extend(extract_entity_names(tree))

print(set(entity_names))
```
